```python
#Implementacion de perceptron con una capa oculta
#con 4 neuronas y una funcion de activacion relu.
#una capa de salida con una neurona y una funcion de activacion
#tangente hiperb√≥lica.
#una capa de entrada de 10 variables.

#usamos api de keras para crear modelos basados en pila de capas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.utils import plot_model
```

```python
#Construccion de modelos.

myModel = Sequential([
    Input(shape=(10, )), #capa de entrada
    Dense(units=4,activation='relu'), #capa oculta A1
    Dense(units=1,activation='tanh') #capa de salida
])
```

```python
myModel.summary()
```

```python
plot_model(myModel, show_shapes = True, to_file="MyModel.png")
```
